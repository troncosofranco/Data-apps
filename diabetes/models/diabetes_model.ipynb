# Diabetes Prediction

## What is Diabetes?
Diabetes is a chronic disease, caused when the pancreas is no longer able to make insulin, hormone of regulate the glucose level in blood. Machine Learning is a powerful tool to Diabetes prediciton based on clinical features

##About this project :
- **Objective**: Classify whether someone has diabetes or not based on clinical features
- **Dataset**: consists of id's and multiple clinical features and one Outcome variable
- **Independent variables:** No. of Patient, Sugar Level Blood, Age, Gender, Creatinine ratio(Cr), Body Mass Index (BMI), Urea, Cholesterol (Chol), Fasting lipid profile, including total, LDL, VLDL, Triglycerides(TG) and HDL Cholesterol , HBA1C, Outcome (the patient's diabetes disease class may be Diabetic (Y) or Non-Diabetic (N).

## Dataset Attributes:
- Gender: Male or Female
- Age: In Years (20-79)
- Urea: in mg/dl (0.5-38.9)
- Creatinine Ratio: Î¼mol/L (48-80)
- Body Mass Index: 19-47
- LDL: mmol/L (0.3-9.9)
- Cholesterol: mmol/L (0.0-10.3)
- VLDL: mmol/L (0.1-35)
- Triglycerides: mmol/L (0.3-13.8)
- HDL: mmol/L (0.2-9.9)
- HBA1C: mmol/L (0.9-16)
- BMI (19 - 47.75)
- Class (outcome): Diabetic or Non-Diabetic

More details in [Diabetes prediction and analysis using medical attributes: A Machine learning approach](https://www.researchgate.net/publication/328766758_Predicting_Diabetes_Mellitus_With_Machine_Learning_Techniques)

##Hyperparameter tuning

Hyperparameter tuning for each model is performed in the current project. The tunin is carried out using Grid Search methodology, which employs a different combination of all the specified hyperparameters and their values, estimating the performance for each combination. Thus, the best value for the model hyperparameters is selected

##Hyperparameter tuning process
The hyperparameter tuning process involves the following stages:
1. Select the model. 
2. Check the parameters of the model.
3. Select the methods for hyperparamenter searching 
4. Select the cross-validation approach
5. Define the score and evaluate the model.
6. Fit the best model.
7. Predict on testing data using the score selected and performance metrics

##Performance metrics
The performance metrics commonly employed in classification machine learning algorithms are: Confusion Matrix, F1 Score, Precision Score, Recall Score

###Confusion matrix
Confusion matrix represents a tabular visualization comparing the model predictions versus the truth labels.

###Recall score
It represents the fraction of positives/negative events that were correctly predicted

###Precision score
Precision Score is the fraction of predicted positives/negatives events that are actually positive/negatives.


###F1 Score
It represents the harmonic mean between precision and recall.


In this project F1 score metric was selected, based on the following items:
- It is a robust metric between Precision and Recall metrics.
- F1 score is a suitable metric when there are imbalanced classes, as it was observed in the previous case.





# 9.1. K Nearest Neighbours: KNN
KNN is a non-parametric algorithm, which classifies data points based on their proximity, and consequently association to other variable data or features

```python
#Import KNN and model selection
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import RepeatedStratifiedKFold

#Import metrics
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.metrics import f1_score, precision_score, recall_score

#Import Grid Search
from sklearn.model_selection import GridSearchCV
```

```python
#List of hyperparameters to tune using BayesianSearchCV

knn= KNeighborsClassifier() #model
n_neighbors = list(range(15,25)) #neighbors number
p=[1,2]
weights = ['uniform', 'distance']
metric = ['euclidean', 'manhattan', 'minkowski']

#Define hyperparameters as dictionary
hyperparameters = dict(n_neighbors=n_neighbors, p=p ,weights=weights , metric=metric)

#Building the model
!pip install scikit-optimize 
from skopt import BayesSearchCV
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
grid_search = GridSearchCV(estimator=knn, param_grid=hyperparameters, n_jobs=-1, cv=cv, scoring='f1',error_score=0)
```

```python
#Fit the model
best_model_knn = grid_search.fit(X_train,y_train)
```

```python
#Print the results of GridSearchCV -> optimal hyperparameters
print('Best leaf_size:', best_model_knn.best_estimator_.get_params()['leaf_size'])
print('Best p:', best_model_knn.best_estimator_.get_params()['p'])
print('Best n_neighbors:', best_model_knn.best_estimator_.get_params()['n_neighbors'])
```

**Conclusion:** According to the Bayesian Optimization, the optimal hyperparameters are:
- Best leaf_size: 30
- Best p: 2
- Best n_neighbors: 23

```python
#Prediction of KNN model
knn_pred = best_model_knn.predict(X_test)
```

```python
#Print the metric summary report
print("Classification Report is:\n",classification_report(y_test,knn_pred))
```

```python
#Print the rest of performance metrics
!pip install matplotlib==3.1.3

print("F1-score:",f1_score(y_test,knn_pred))
print("\n Precision score: ",precision_score(y_test,knn_pred))
print("\n Recall score:",recall_score(y_test,knn_pred))
print("\n Confusion Matrix:\n")

plt.figure(figsize = (10,7))

sns.heatmap(confusion_matrix(y_test,knn_pred), square=True, annot=True, cmap='Blues', fmt='d', cbar=True)
```

# 9.2. Naive Bayes

Naive Bayes algorithm is widely employed in stream classification system under production environment. This algorithm adopts the principle class conditional independence from the Bayes Theorem. In this way, the effect of one feature is not sensitive by other one in the probability of the outcome. Furthermore, each feature presents an equal effect on the outcome.


```python
#Import libraries
from sklearn.naive_bayes import GaussianNB
#from sklearn.model_selection import GridSearchCV

#bayer_search = BayesSearchCV(estimator=knn, search_spaces=hyperparameters, n_jobs=-1, cv=cv, scoring='f1',error_score=0)
#Define the grid model of Naive Bayes
param_grid_nb = {'var_smoothing': np.logspace(0,-2, num=100)}
nbModel_grid = BayesSearchCV(estimator=GaussianNB(), search_spaces = param_grid_nb, verbose=1, cv=10, n_jobs=-1)
```

```python
#Determine best hyperparameter for NB model
best_model_nb = nbModel_grid.fit(X_train, y_train)
```

```python
#Make prediction
nb_pred = best_model_nb.predict(X_test)
```

```python
#Print the metric summary report
print("Classification Report is:\n",classification_report(y_test,nb_pred))
```

```python
#Print the rest of performance metrics
print("\n F1-Score:",f1_score(y_test,nb_pred))
print("\n Precision score:",precision_score(y_test,nb_pred))
print("\n Recall score:",recall_score(y_test,nb_pred))
print("\n Confusion Matrix:\n")

plt.figure(figsize = (10,7))
sns.heatmap(confusion_matrix(y_test,nb_pred), square=True, annot=True, cmap='Blues', fmt='d', cbar=True)
```

#9.3 Support Vector Machine (SVM)

SVM algorithm is widely employed for classification problems. SVM constructs a hyperplane, also called decision boundary, where the distance between two classes is maximum. 


```python
#Import libraries
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
```

```python
#Define model
model = SVC()

#Define hyperparameters
kernel = ['poly', 'rbf', 'sigmoid']
C = [50, 10, 1.0, 0.1, 0.01]
gamma = ['scale']

#Hyperparameters to dictionary
grid = dict(kernel=kernel, C=C, gamma=gamma)
```

```python
#Find the optimal hyperparameters of the model using Grid Search CV methodology
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
grid_search = BayesSearchCV(estimator=model, search_spaces=grid, n_jobs=-1, cv=cv, scoring='f1',error_score=0)
```

```python
#Fit the model and make pred
grid_result = grid_search.fit(X_train, y_train)
svm_pred = grid_result.predict(X_test)
```

```python
#Print the Classification report using SVM
print("Classification Report:\n",classification_report(y_test,svm_pred))
```

```python
#Print the Performance Metrics
print("\n F1-Score:",f1_score(y_test,svm_pred))
print("\n Precision score:",precision_score(y_test,svm_pred))
print("\n Recall score:",recall_score(y_test,svm_pred))
print("\n Confusion Matrix:\n")

plt.figure(figsize = (10,7))
sns.heatmap(confusion_matrix(y_test,svm_pred), square=True, annot=True, cmap='Blues', fmt='d', cbar=True)
```
